## linestyle (ls)

**Por defecto es una línea sólida ('-')**

dotted (punteado): línea punteada.

dashed (guiones): línea discontinua.

dashdot (punto-guion): línea con una secuencia de puntos y guiones alternados.

solid (sólido): línea continua.

None o '' (ninguno): Indica que no hay línea, solo marcadores o puntos en el gráfico.

' ' (espacio en blanco): Indica que no hay línea ni marcadores en el gráfico.

ls='-': Línea sólida 

ls='--': Línea discontinua 

ls=':': Línea punteada 

ls='-.': Línea punto-guion

## plt.legend

Para agregar leyendas al gráfico.

El parámetro loc en legend() especifica dónde debe ubicarse la leyenda dentro del área del gráfico.

`best`: Matplotlib elige la mejor ubicación automáticamente (minimizando la interferencia con los datos del gráfico).

`upper right`: Coloca la leyenda en la esquina superior derecha del gráfico.

`upper left`: Coloca la leyenda en la esquina superior izquierda.

`lower left`: Coloca la leyenda en la esquina inferior izquierda.

`lower right`: Coloca la leyenda en la esquina inferior derecha.

`right`: Coloca la leyenda a la derecha del gráfico.

`center left`: Coloca la leyenda en el centro a la izquierda.

`center right`: Coloca la leyenda en el centro a la derecha.

`lower center`: Coloca la leyenda en la parte inferior al centro.

`upper center`: Coloca la leyenda en la parte superior al centro.

`center`: Coloca la leyenda en el centro del gráfico.

---

**fontsize**: controla el tamaño de la fuente de los textos dentro de la leyenda (`small`, `medium`, `large`).

**frameon**: es un booleano que controla si la leyenda debe tener un marco alrededor. **(defecto, es True)**.

**title**: permite agregar un título al cuadro de la leyenda.

**borderpad**: establece el relleno entre el borde del cuadro de la leyenda y los contenidos de la leyenda. 

**labelspacing**: controla el espacio vertical entre las líneas de la leyenda.

**handletextpad**: controla el espacio entre el identificador y el texto en la leyenda.

**borderaxespad**: establece el espacio entre el cuadro de la leyenda y los ejes del gráfico

```python
a = np.arange(0,11,1)
b = np.arange(0,21,2)
c = np.arange(0,31,3)
plt.figure(figsize=(8, 6))
plt.plot(a, b, c = 'navy', ls = 'dotted', lw = 2, marker = 'o', ms = 4, mec='red', mfc = 'yellow', label = 'Gráfico 1')
plt.plot(a, c, c = 'navy', lw = 2, marker = 'o', ms = 4, mec='red', mfc = 'black', label = 'Gráfico 2')
plt.legend()
plt.show()
```

```python
a = np.arange(0, 11, 1)
b = np.arange(0, 21, 2)
c = np.arange(0, 31, 3)
plt.figure(figsize=(8, 6))
plt.plot(a, b, color='navy', linestyle='dotted', linewidth=2, marker='o', markersize=4, markeredgecolor='red', markerfacecolor='yellow', label='Gráfico 1')
plt.plot(a, c, color='navy', linewidth=2, marker='o', markersize=4, markeredgecolor='red', markerfacecolor='black', label='Gráfico 2')
plt.legend(loc='upper center', fontsize='medium', frameon=True, title='Leyenda', borderpad=1, labelspacing=1, handletextpad=2, borderaxespad=1)
plt.show()
```

## plt.tick_params

Permite personalizar la apariencia de los ticks y las etiquetas de los ejes en tus gráficos.

axis: Especifica sobre qué eje se aplicarán los cambios. Los valores pueden ser 'x', 'y', o 'both' (**defecto es 'both'**).

which: Puede ser 'major', 'minor', o 'both', especificando a qué ticks se aplicarán los cambios. **defecto, es 'major'**.

direction: Puede ser 'in', 'out', o 'inout'. Esto controla la dirección en la que los ticks apuntan desde el eje. **defecto, es 'out'**.

length: Longitud de los ticks en puntos; **defecto es 4**.

width: Ancho de los ticks en puntos; **defecto es 0.5**.

color: Color de los ticks; **defecto es 'black'**.

pad: Distancia en puntos entre el tick y la etiqueta correspondiente; **defecto es 3.5**.

labelsize: Tamaño de la fuente de las etiquetas de los ticks; **defecto es None**.

labelcolor: Color de las etiquetas de los ticks **defecto es None**.

bottom, top, left, right: Booleanos para indicar si los ticks deberían ser visibles en los lados inferior, superior, izquierdo o derecho del gráfico. 

**defecto, están configurados en True para bottom y left y False para top y right**.

labelbottom, labeltop, labelleft, labelright: Booleanos para indicar si las etiquetas de los ticks deberían ser visibles en los lados especificados. 

**defecto, configurados en True para labelbottom y labelleft y False para labeltop y labelright**.

```python
import matplotlib.pyplot as plt
import numpy as np

a = np.arange(0, 11, 1)
b = np.arange(0, 21, 2)
c = np.arange(0, 31, 3)

plt.figure(figsize=(8, 6))
plt.grid(True)

plt.plot(a, b, color='navy', linestyle='dotted', linewidth=2, marker='o', markersize=4, markeredgecolor='red', markerfacecolor='yellow', label='Gráfico 1')
plt.plot(a, c, color='navy', linewidth=2, marker='o', markersize=4, markeredgecolor='red', markerfacecolor='black', label='Gráfico 2')

plt.grid(color='black', linestyle='--', linewidth=0.5)

font_title = {'family': 'monospace', 'color': 'black', 'size': 20}
plt.title("Título del Gráfico", fontdict=font_title, loc='center', pad=10)
font_label = {'family': 'fantasy', 'color': 'teal', 'size': 15}
plt.xlabel('Eje X', fontdict=font_label)
plt.ylabel('Eje Y', fontdict=font_label)

plt.xlim(2, 8)
plt.ylim(3, 15)

plt.xticks([2, 4, 6, 8])
plt.yticks([3, 6, 9, 12, 15])

plt.tick_params(axis='x', direction='out', length=6, width=2, colors='blue', grid_color='blue', grid_alpha=0.5)
plt.tick_params(axis='y', direction='out', length=6, width=2, colors='green', grid_color='green', grid_alpha=0.5)

plt.legend()
plt.show()
```

```python
a = np.arange(0, 11, 1)
b = np.arange(0, 21, 2)
c = np.arange(0, 31, 3)

plt.figure(figsize=(8, 6))
plt.grid(True)

plt.plot(a, b, color='darkgreen', linestyle='-', linewidth=2, marker='s', markersize=6, markeredgecolor='black', markerfacecolor='lime', label='Gráfico 1')
plt.plot(a, c, color='darkred', linestyle='--', linewidth=2, marker='d', markersize=6, markeredgecolor='black', markerfacecolor='pink', label='Gráfico 2')

plt.grid(color='gray', linestyle='-', linewidth=0.5)

font_title = {'family': 'serif', 'color': 'darkred', 'size': 20}
plt.title("Variación de los Datos", fontdict=font_title, loc='center', pad=20)
font_label = {'family': 'sans-serif', 'color': 'darkblue', 'size': 15}
plt.xlabel('Eje X', fontdict=font_label)
plt.ylabel('Eje Y', fontdict=font_label)

plt.xlim(2, 8)
plt.ylim(3, 30)

plt.xticks([2, 4, 6, 8], labels=['Inicio', 'Medio', 'Avanzado', 'Final'], rotation=45)
plt.yticks([3, 10, 17, 24, 30])

plt.tick_params(axis='x', direction='in', length=10, width=1, colors='navy', grid_color='navy', grid_alpha=0.7)
plt.tick_params(axis='y', direction='in', length=10, width=1, colors='maroon', grid_color='maroon', grid_alpha=0.7)

plt.legend(loc='upper left')
plt.show()
```

## plt.xticks y plt.yticks()

Establece las marcas del eje x / eje y

```python
a = np.arange(0,11,1)
b = np.arange(0,21,2)
c = np.arange(0,31,3)
plt.figure(figsize=(8, 6))
plt.grid(True)
plt.plot(a, b, c = 'navy', ls = 'dotted', lw = 2, marker = 'o', ms = 4, mec='red', mfc = 'yellow', label = 'Gráfico 1')
plt.plot(a, c, c = 'navy', lw = 2, marker = 'o', ms = 4, mec='red', mfc = 'black', label = 'Gráfico 2')
plt.grid(color='black', linestyle='--', linewidth=0.5)
font_title = {'family': 'monospace', 'color':'black','size':20}
plt.title("título del gráfico", fontdict = font_title, loc = 'center', pad=10)
font_label = {'family': 'fantasy', 'color':'teal','size':15}
plt.xlabel('Eje x', fontdict=font_label)
plt.ylabel('Eje y', fontdict=font_label)
plt.xlim(2,8)
plt.ylim(3,15)
plt.xticks([2, 4, 6, 8])  
plt.yticks([3, 6, 9, 12, 15])  
plt.legend()
plt.show()
```
