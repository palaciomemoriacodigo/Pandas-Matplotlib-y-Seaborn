# Importar librerias

```python
# pip install matplotlib
# pip install matplotlib
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
print(matplotlib.__version__)
```

```python
df = sns.load_dataset('titanic').dropna(subset=['age', 'fare', 'survived'])
tips = sns.load_dataset('tips')
```

# Tipos de grÃ¡ficos

## GrÃ¡fico de lÃ­neas

Variables temporales o secuenciales (series de tiempo).

Relaciones entre dos variables continuas.

```python
edad_media_clase = df.groupby('class',observed=True)['age'].mean()
plt.figure(figsize=(10, 5))
ax = plt.subplot(111)
plt.plot(edad_media_clase.index, edad_media_clase.values, marker='o', linestyle='--', color='r')
plt.title('Edad media por clase')
plt.xlabel('Clase de cabina')
plt.ylabel('Edad media')
plt.grid(True)
plt.show()
```

```python
edad_media_clase = df.groupby('class', observed=True)['age'].mean()
ax = edad_media_clase.plot(marker='o', linestyle='--', color='r', figsize=(10, 5), title='Edad media por clase')
ax.set_xlabel('Clase de cabina')
ax.set_ylabel('Edad media')
ax.grid(True)
plt.show()
```

```python
tarifa_promedio = df.groupby('class', observed=True)['fare'].mean()
plt.figure(figsize=(8, 6))
ax = plt.subplot(111)
plt.grid(True)
font_title = {'family': 'monospace', 'color': 'darkred', 'size': 20}
plt.title('Tarifa/Clase', fontdict=font_title, loc='center', pad=10)
font_label = {'family': 'fantasy', 'color': 'teal', 'size': 15}
plt.xlabel('Clase', fontdict=font_label)
plt.ylabel('Tarifa Promedio ($)', fontdict=font_label)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.plot(tarifa_promedio.index,tarifa_promedio, marker='o', color='navy', ls='dotted', lw=2, mec='red', mfc='yellow', markersize=8)
plt.legend(["Tarifa Promedio"])
plt.show()
```

```python
tarifa_promedio = df.groupby('class', observed=True)['fare'].mean()
ax = tarifa_promedio.plot(marker='o', color='navy', ls='dotted', lw=2, mec='red', mfc='yellow', markersize=8, figsize=(8, 6), grid=True, legend=True)
ax.set_title('Tarifa/Clase', fontdict={'family': 'monospace', 'color': 'darkred', 'size': 20}, loc='center', pad=10)
ax.set_xlabel('Clase', fontdict={'family': 'fantasy', 'color': 'teal', 'size': 15})
ax.set_ylabel('Tarifa Promedio ($)', fontdict={'family': 'fantasy', 'color': 'teal', 'size': 15})
ax.grid(color='gray', linestyle='--', linewidth=0.5)
plt.show()
```

```python
tarifa_promedio = df.groupby('embark_town', observed=True)['fare'].mean()

plt.figure(figsize=(10, 6))
ax = plt.subplot(111) 
ax.set_facecolor('whitesmoke')  
plt.grid(True, color='grey', linestyle=':', linewidth=0.5, alpha=0.5)
font_title = {'family': 'Arial', 'color': 'darkblue', 'size': 20, 'weight': 'bold'}
plt.title('Tarifa/Ciudad de Embarque', fontdict=font_title, loc='center', pad=20)
font_label = {'family': 'Arial', 'color': 'darkslategray', 'size': 15, 'weight': 'bold'}
plt.xlabel('Ciudad de Embarque', fontdict=font_label)
plt.ylabel('Tarifa Promedio ($)', fontdict=font_label)
plt.plot(tarifa_promedio.index,tarifa_promedio, marker='o', color='royalblue', ls='-', lw=2, mec='crimson', mfc='lightblue', markersize=10)
plt.legend(["Tarifa Promedio"], frameon=True, framealpha=0.8, facecolor='lightgrey', edgecolor='black')
plt.xticks(rotation=45) 
ax.spines['top'].set_visible(False)  
ax.spines['right'].set_visible(False)  
plt.show()
```

```python
tarifa_promedio = df.groupby('embark_town', observed=True)['fare'].mean()
ax = tarifa_promedio.plot(marker='o', color='royalblue', ls='-', lw=2, mec='crimson', mfc='lightblue', markersize=10, figsize=(10, 6), grid=True, legend=True, rot=45)
ax.set_facecolor('whitesmoke')
ax.set_title('Tarifa/Ciudad de Embarque', fontdict={'family': 'Arial', 'color': 'darkblue', 'size': 20, 'weight': 'bold'}, loc='center', pad=20)
ax.set_xlabel('Ciudad de Embarque', fontdict={'family': 'Arial', 'color': 'darkslategray', 'size': 15, 'weight': 'bold'})
ax.set_ylabel('Tarifa Promedio ($)', fontdict={'family': 'Arial', 'color': 'darkslategray', 'size': 15, 'weight': 'bold'})
ax.grid(True, color='grey', linestyle=':', linewidth=0.5, alpha=0.5)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.show()
```

## GrÃ¡fico de Barras Horizontales

Similar al grÃ¡fico de barras, pero mÃ¡s adecuado cuando tienes muchas categorÃ­as o cuando quieres que las etiquetas del eje x sean mÃ¡s legibles

```python
pasajeros_sexo = df['sex'].value_counts()
plt.figure(figsize=(10, 5))
ax = plt.subplot(111)
plt.barh(pasajeros_sexo.index, pasajeros_sexo.values, color='c')
plt.title('NÃºmero de pasajeros por sexo')
plt.xlabel('NÃºmero de pasajeros')
plt.ylabel('Sexo')
plt.show()
```

```python
pasajeros_sexo = df['sex'].value_counts()
ax = pasajeros_sexo.plot(kind='barh', color='c', figsize=(10, 5), title='NÃºmero de pasajeros por sexo')
ax.set_xlabel('NÃºmero de pasajeros')
ax.set_ylabel('Sexo')
plt.show()
```

```python
supervivientes_por_sexo = df.pivot_table(index='sex', columns='survived', aggfunc='size')
categorias = supervivientes_por_sexo.index
no_sobrevivieron = supervivientes_por_sexo[0]  
sobrevivieron = supervivientes_por_sexo[1]     

y = np.arange(len(categorias))

plt.figure(figsize=(10, 5))
ax = plt.subplot(111)

plt.barh(y, no_sobrevivieron, color='black', label='No Sobrevivieron')
plt.barh(y, sobrevivieron, left=no_sobrevivieron, color='#66b3ff', label='Sobrevivieron')

ax.set_facecolor('whitesmoke')
ax.set_title('Supervivientes por Sexo en el Titanic', fontsize=16, color='darkblue', weight='bold')
ax.set_xlabel('NÃºmero de Pasajeros', fontsize=14, color='darkgreen')
ax.set_ylabel('Sexo', fontsize=14, color='darkgreen')
ax.grid(color='grey', linestyle=':', linewidth=0.5, alpha=0.5)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.set_yticks(y)
ax.set_yticklabels(categorias)
ax.legend(title="Estado", title_fontsize='13')

plt.show()
```

```python
supervivientes_por_sexo = df.pivot_table(index='sex', columns='survived', aggfunc='size')
supervivientes_por_sexo.plot(kind='barh', stacked=True, color=['black', '#66b3ff'], figsize=(10, 5))
ax = plt.gca()
ax.set_facecolor('whitesmoke')
ax.set_title('Supervivientes por Sexo en el Titanic', fontsize=16, color='darkblue', weight='bold')
ax.set_xlabel('NÃºmero de Pasajeros', fontsize=14, color='darkgreen')
ax.set_ylabel('Sexo', fontsize=14, color='darkgreen')
ax.grid(color='grey', linestyle=':', linewidth=0.5, alpha=0.5)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.legend(['No Sobrevivieron', 'Sobrevivieron'], title="Estado", title_fontsize='13')
plt.show()
```

## GrÃ¡fico de DispersiÃ³n (Scatter Plot)

Relaciones entre dos variables continuas.

Identificar patrones o tendencias.

Detectar la presencia de valores atÃ­picos o anomalÃ­as.

```python
plt.figure(figsize=(10, 6))
ax = plt.subplot(111)
plt.scatter(df['age'], df['fare'], alpha=0.7, color='darkred', edgecolor='black')
ax.set_facecolor('whitesmoke')
ax.set_title('RelaciÃ³n entre Edad y Tarifa en el Titanic', fontsize=16, color='darkblue', weight='bold')
ax.set_xlabel('Edad', fontsize=14, color='darkgreen')
ax.set_ylabel('Tarifa', fontsize=14, color='darkgreen')
ax.grid(color='grey', linestyle='--', linewidth=0.5, alpha=0.5)
plt.show()
```

```python
ax = df.plot(kind='scatter', x='age', y='fare', color='darkred', edgecolor='black', alpha=0.7, figsize=(10, 6), title='RelaciÃ³n entre Edad y Tarifa en el Titanic')
ax.set_facecolor('whitesmoke')
ax.set_xlabel('Edad', fontsize=14, color='darkgreen')
ax.set_ylabel('Tarifa', fontsize=14, color='darkgreen')
ax.grid(color='grey', linestyle='--', linewidth=0.5, alpha=0.5)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
ax = plt.subplot(111)
class_map = {'First': 1, 'Second': 2, 'Third': 3}
df['class_numeric'] = df['class'].map(class_map)
plt.scatter(df['class_numeric'], df['fare'], alpha=0.7, color='gold', edgecolor='black')
ax.set_facecolor('whitesmoke')
ax.set_title('RelaciÃ³n entre Clase y Tarifa en el Titanic', fontsize=16, color='darkblue', weight='bold')
ax.set_xlabel('Clase', fontsize=14, color='darkgreen')
ax.set_ylabel('Tarifa', fontsize=14, color='darkgreen')
ax.grid(color='grey', linestyle='--', linewidth=0.5, alpha=0.5)
ax.set_xticks([1, 2, 3])
ax.set_xticklabels(['Primera', 'Segunda', 'Tercera'])

plt.show()

```

```python
ax = df.plot(kind='scatter', x='class', y='fare', color='gold', edgecolor='black', alpha=0.7, figsize=(10, 6), title='RelaciÃ³n entre Clase y Tarifa en el Titanic')
ax.set_facecolor('whitesmoke')
ax.set_xlabel('Clase', fontsize=14, color='darkgreen')
ax.set_ylabel('Tarifa', fontsize=14, color='darkgreen')
ax.grid(color='grey', linestyle='--', linewidth=0.5, alpha=0.5)
ax.set_xticks([0, 1, 2])
ax.set_xticklabels(['Primera', 'Segunda', 'Tercera'])
plt.show()
```

## GrÃ¡fico de Caja (Box Plot)

Variables numÃ©ricas.

Comparar distribuciones de una o mÃ¡s variables.

Identificar valores atÃ­picos, dispersiÃ³n y simetrÃ­a en los datos.

```python
plt.figure(figsize=(8, 6))
data_to_plot = df.groupby('class', observed=True)['fare'].apply(list)
plt.boxplot(data_to_plot, labels=data_to_plot.index)
font_title = {'family': 'monospace', 'color':'darkblue','size':15}
font_label = {'family': 'fantasy', 'color':'teal','size':15}
plt.title('DistribuciÃ³n de Tarifas por Clase en el Titanic',fontdict=font_title)
plt.xlabel('Clase',fontdict=font_label)
plt.ylabel('Tarifa',fontdict=font_label)
plt.show()
```

```python
ax = df.boxplot(column='fare', by='class', figsize=(8, 6))
ax.set_title('DistribuciÃ³n de Tarifas por Clase en el Titanic')
ax.figure.suptitle('')
ax.set_xlabel('Clase')
ax.set_ylabel('Tarifa')
ax.grid(color='grey', linestyle='--', linewidth=0.5, alpha=0.5)
ax.title.set_fontsize(15)
ax.title.set_color('darkblue')
ax.xaxis.label.set_fontsize(15)
ax.xaxis.label.set_color('teal')
ax.yaxis.label.set_fontsize(15)
ax.yaxis.label.set_color('teal')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
data_to_plot = df.groupby('sex', observed=True)['fare'].apply(list)
plt.boxplot(data_to_plot, labels=data_to_plot.index)
font_title = {'family': 'monospace', 'color':'darkblue','size':15}
font_label = {'family': 'fantasy', 'color':'teal','size':15}
plt.title('DistribuciÃ³n de Tarifas por Clase en el Titanic',fontdict=font_title)
plt.xlabel('Clase',fontdict=font_label)
plt.ylabel('Tarifa',fontdict=font_label)
plt.show()
```

```python
ax = df.boxplot(column='fare', by='sex', figsize=(8, 6))
ax.set_title('DistribuciÃ³n de Tarifas por Sexo en el Titanic')
ax.figure.suptitle('')
ax.set_xlabel('Sexo')
ax.set_ylabel('Tarifa')
ax.grid(color='grey', linestyle='--', linewidth=0.5, alpha=0.5)
ax.title.set_fontsize(15)
ax.title.set_color('darkblue')
ax.xaxis.label.set_fontsize(15)
ax.xaxis.label.set_color('teal')
ax.yaxis.label.set_fontsize(15)
ax.yaxis.label.set_color('teal')
plt.show()
```
