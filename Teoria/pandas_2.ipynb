# pd.to_numeric()

Convierte argumentos a números

errors = 'coerce' --> podemos convertir los valores no numéricos a NaN para asegurar que toda la columna sea numérica.

```python
titanic_df['fare'] = pd.to_numeric(titanic_df['fare'], errors='coerce')
```

# pd.to_datetime()

Convierte argumentos a formato de fecha y hora.

format= --> Define el formato de la fecha en la columna original.

```python
dowjones_df
```

```python
dowjones_df.info()
```

```python
dowjones_df['Date'] = pd.to_datetime(dowjones_df['Date'], format='%Y-%m-%d')
```

```python
dowjones_df.head(2)
```

```python
dowjones_df['Date'] = pd.to_datetime(dowjones_df['Date'], format='%Y-%m-%d')
```

```python
dowjones_df['Year'] = dowjones_df['Date'].dt.year
dowjones_df['Month'] = dowjones_df['Date'].dt.month
dowjones_df['Day'] = dowjones_df['Date'].dt.day
dowjones_df['Day_of_week'] = dowjones_df['Date'].dt.day_of_week
dowjones_df['Day_of_year'] = dowjones_df['Date'].dt.day_of_year
dowjones_df['Days_in_month'] = dowjones_df['Date'].dt.days_in_month
dowjones_df['Days_in_month'] = dowjones_df['Date'].dt.weekday
```
