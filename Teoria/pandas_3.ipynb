## Filtrado Condicional Múltiple

```python
filtrado_condicional_multiple = titanic_df[(titanic_df['age'] > 30) & (titanic_df['pclass'] == 3) & (titanic_df['survived'] == 1)]
filtrado_condicional_multiple
```

```python
filtrado_condicional_multiple = titanic_df[((titanic_df['age'] > 30) &  (titanic_df['survived'] == 1)) | ((titanic_df['pclass'] == 3) & (titanic_df['survived'] == 0))]
filtrado_condicional_multiple
```

## equals()

Compara si 2 DataFrames son iguales (comparar conjuntos de datos con cambios aplicados)

```python
df1 = titanic_df.head(3)
df2 = titanic_df.head(3)
print(df1.equals(df2)) 
```

```python
class_equals_survived = df1['pclass'] == df2['pclass']
print(class_equals_survived.head())
```

```python
df1 = titanic_df.head(3)
df2 = titanic_df.tail(3)
print(df1.equals(df2)) 
```

# Operaciones de búsqueda y consulta

## pd.isna() y pd.notna()

pd.isna() genera una Serie booleana que indica dónde hay valores NaN dentro de un DataFrame o Serie. 

pd.notna() hace lo contrario.

```python
edad_nan = pd.isna(titanic_df['age'])
print(edad_nan.head())
titanic_df[edad_nan]
```

```python
edad_sin_nan = pd.notna(titanic_df['age'])
print(edad_sin_nan.head())
titanic_df[edad_sin_nan]

```

## query()

Permite filtrar un DataFrame usando una consulta (parecido a SQL).

```python
query = "pclass == 1 and fare > 100"
primera_clase_caro = titanic_df.query(query)
primera_clase_caro.head()
```

```python
query = "pclass == 1 and sex == 'female'"
mujeres_primera_clase = titanic_df.query(query)
mujeres_primera_clase
```

```python
mujeres_primera_clase.describe()
```

```python
query = "pclass == 3 and sex == 'female'"
mujeres_tercera_clase = titanic_df.query(query)
mujeres_tercera_clase
```

```python
mujeres_tercera_clase.describe()
```

## .where() y .filter()

Con el where:

Conservará la misma estructura de índice que el DataFrame original, donde no se encuentre coincidencia se sustituirá por NaN. (mantenemos el tamaño del DataFrame para comparaciones posteriores).

Con el filter:

Seleccionamos columnas o filas basadas en especificaciones de sus etiquetas.

```python
tercera_edad = titanic_df['age'].where(titanic_df['age'] >= 65)
tercera_edad
```

```python
filtrar_columnas = titanic_df.filter(like='class')
print(filtrar_columnas)
```

 # Operaciones de filtrado avanzadas

##  Máscaras booleanas / mask 

Son Series o DataFrames donde cada elemento es un booleano.

&  --> AND

| --> OR 

~ --> NOT

```python

tercera_edad = titanic_df[titanic_df['age'] > 65]
tercera_edad
```

```python
mask = (titanic_df['age'] > 65) & (titanic_df['pclass'].isin([2, 3]))
tercera_edad_no_rico = titanic_df[mask]
tercera_edad_no_rico
```

```python
mask = (titanic_df['age'] > 65) & (titanic_df['pclass'].isin([1]))
no_sea_tercera_edad_y_rico = titanic_df[~mask]
no_sea_tercera_edad_y_rico
```
