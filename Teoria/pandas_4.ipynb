```python
diccionario_pokemon = {
    'Nombre': ['Pikachu', 'Charmander', 'Bulbasaur', 'Squirtle', 'Jigglypuff'],
    'Tipo': ['Eléctrico', 'Fuego', 'Planta', 'Agua', 'Normal'],
    'Nivel': [25, 16, 12, np.NaN, 20],
    'Capturado': [True, False, True, True, False]
}

df_pokemon = pd.DataFrame(diccionario_pokemon)
df_pokemon
```

# to_csv()

index: Incluir o no el índice como columna en el archivo.

header: Si se deben incluir los nombres de las columnas.

columns: Especificar un subconjunto de columnas a escribir.

na_rep: Representación de valores NaN.

```python
df_pokemon.to_csv('pokemon_index_false.csv', index=False)
```

```python
df_pokemon.to_csv('pokemon_index_true.csv', index=True)
```

```python
df_pokemon.to_csv('pokemon_index_false_sin_header.csv', index=False, header=None)
```

```python
df_pokemon.to_csv('pokemon_index_false_2col.csv', index=False, columns = ['Nombre','Tipo'])
```

```python
df_pokemon.to_csv('pokemon_index_true_na.csv', index=True, na_rep='Chapuza')
```
