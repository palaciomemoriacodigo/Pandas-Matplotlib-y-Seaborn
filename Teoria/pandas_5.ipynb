# importación de librerias y datasets

```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
titanic_df = sns.load_dataset('titanic')
tips_df = sns.load_dataset('tips')
```

# groupby

El método groupby permite  agrupar datos en base a una o más columnas.

```python
titanic_df.head()
```

```python
titanic_df.describe()
```

```python
dinero_por_clase = titanic_df.groupby('pclass')['fare'].sum()
print(type(dinero_por_clase))
dinero_por_clase
```

```python
tarifa_media_por_clase = titanic_df.groupby('pclass')['fare'].mean()
print(tarifa_media_por_clase)
```

```python
tarifa_media_por_clase = titanic_df.groupby('pclass')['fare'].median()
print(tarifa_media_por_clase)
```

```python
tips_df.head(3)
```

```python
tips_groupby = tips_df.groupby('smoker',observed=True)['tip'].median()
tips_groupby
```

```python
tips_groupby = tips_df.groupby('smoker',observed=True)['tip'].sum()
tips_groupby
```

```python
dinero_por_clase_y_genero = titanic_df.groupby(['sex', 'pclass'])['fare'].median()
print(type(dinero_por_clase_y_genero))
dinero_por_clase_y_genero
```

## agg()

Aplicar diferentes funciones a diferentes columnas

```python
groupby_apply = titanic_df.groupby('pclass').agg({
    'fare': ['mean', 'std', 'max', 'min'],  
    'age': ['mean', 'median', 'count'],     
    'survived': 'sum'                       
})

print(type(groupby_apply))
groupby_apply
```

```python
tips_df.info()
```

```python
groupby_apply = tips_df.groupby('day',observed=True).agg({
    'tip': ['mean', 'std', 'max', 'min'],  
    'size': ['mean', 'median', 'count']                           
})

print(type(groupby_apply))
groupby_apply
```

## funciones lambda

Filtrar grupos basados en condiciones personalizadas

```python
filtered_groupby = titanic_df.groupby('pclass').filter(lambda x: x['fare'].mean() > 30)
filtered_groupby
```

```python
filtered_groupby['pclass'].value_counts()
```

```python
filtered_survival = titanic_df.groupby('pclass').filter(lambda x: (x['survived'].sum() / len(x)) > 0.4)
filtered_survival
```

```python
filtered_survival['pclass'].value_counts()
```

```python
survived_groupby = titanic_df.groupby('pclass')['survived'].mean()
print(survived_groupby)
```

# pivot_table

Resume y reorganiza los datos de una tabla de entrada, agrupándolos en un formato tabular.

```python
pivot_table_titanic = titanic_df.pivot_table(values='survived', index='pclass', columns='sex', aggfunc='mean')
print(type(pivot_table_titanic))
pivot_table_titanic
```

```python
pivot_table_titanic = titanic_df.pivot_table(values='survived', index='pclass', columns='sex', aggfunc='mean',margins=True, margins_name='Totales')
pivot_table_titanic
```

```python
titanic_df.describe()
```

```python
pivot_table_titanic = titanic_df.pivot_table(values='fare', index='pclass', columns='sex', aggfunc=['mean', 'max', 'min'])
pivot_table_titanic
```

```python
pivot_table_titanic = titanic_df.pivot_table(
    values=['survived', 'fare', 'age'], 
    index='pclass', 
    columns='sex', 
    aggfunc={'survived': 'mean', 'fare': 'median', 'age': 'mean'}
)
pivot_table_titanic
```

# groupby y pivot_table juntos

```python
agrupado = titanic_df.groupby(['pclass', 'sex']).agg({'fare': 'mean'}).reset_index()
agrupado
```

```python
pivot_table = agrupado.pivot_table(values='fare', index='pclass', columns='sex')
pivot_table
```

```python
agrupado = titanic_df.groupby(['pclass', 'sex']).agg({'fare': 'mean'}).reset_index()
pivot_table = agrupado.pivot_table(values='fare', index='pclass', columns='sex')
print(type(pivot_table))
print(pivot_table.shape)
pivot_table
```
