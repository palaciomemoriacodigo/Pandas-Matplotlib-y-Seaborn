# Importación librerias y creación de datasets

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import pandas as pd

data_entrenadores = {
    'Entrenador_ID': [1, 2, 3, 4, 5],
    'Nombre': ['Ash', 'Misty', 'Brock', 'Lance', 'Scar'],
    'Región': ['Kanto', 'Kanto', 'Kanto', 'Johto','Johto']
}

df_entrenadores = pd.DataFrame(data_entrenadores)
df_entrenadores
```

```python
data_pokemon = {
    'Entrenador_ID': [1, 1, 2, 2, 3, 4, 4, 6],
    'Pokemon': ['Pikachu', 'Bulbasaur', 'Starmie', 'Onix', 'Geodude', 'Dragonite', 'Charizard','Charmander']
}

df_pokemon = pd.DataFrame(data_pokemon)
df_pokemon
```

# Merge

## Inner join

 El `Inner Join` devuelve solo aquellos registros que tienen claves coincidentes en ambos DataFrames.

```python
inner_join_df = pd.merge(df_entrenadores, df_pokemon, on='Entrenador_ID', how='inner')
inner_join_df
```

## Left Join

El `Left Join` devuelve todos los registros del DataFrame de la izquierda y los registros coincidentes del DataFrame de la derecha

```python
left_join_df = pd.merge(df_entrenadores, df_pokemon, on='Entrenador_ID', how='left')
left_join_df
```

 ## Right Join
 
 El `Right Join` devuelve todos los registros del DataFrame de la derecha y los registros coincidentes del DataFrame de la izquierda.

```python
left_join_df = pd.merge(df_entrenadores, df_pokemon, on='Entrenador_ID', how='right')
left_join_df
```

## Outer Join

 El `Outer Join` combina los resultados de ambos left y right joins.

```python
outer_join_df = pd.merge(df_entrenadores, df_pokemon, on='Entrenador_ID', how='outer')
outer_join_df
```

# Concat

La función `concat` concatena objetos de pandas a lo largo de un eje particular 

axis=0: filas
 
axis=1: columnas

```python
df_estadisticas = pd.DataFrame({
'Pokemon': ['Pikachu', 'Charmander'],
'HP': [37, 38]
})

df_tipos = pd.DataFrame({
'Pokemon': ['Bulbasaur', 'Squirtle'],
'Type': ['Grass', 'Water']
})

df_concatenado = pd.concat([df_estadisticas, df_tipos], axis=1)
df_concatenado
```

```python
df_estadisticas = pd.DataFrame({
'Pokemon': ['Pikachu', 'Charmander'],
'HP': [37, 38]
})

df_tipos = pd.DataFrame({
'Pokemon': ['Bulbasaur', 'Squirtle'],
'Type': ['Grass', 'Water']
})

df_concatenado = pd.concat([df_estadisticas, df_tipos], axis=0)
df_concatenado
```

# Join

La función `join` permite realizar la unión de DataFrames a partir de los índices 

```python
import pandas as pd

df_estadisticas = pd.DataFrame({
    'PS': [35, 37, 39, 44, 42],  
    'Ataque': [55, 56, 52, 48, 50]
}, index=['Pikachu','Pikachu', 'Charmander', 'Squirtle', 'Bulbasur'])

df_habilidades = pd.DataFrame({
    'Habilidad Primaria': ['Estática', 'Cola rayo', 'Mar de Llamas', 'Torrente','Rayo Hielo'],
    'Habilidad Secundaria': ['Pararrayos', 'Impactrueno' , 'Poder Solar', 'Lluvia', 'Viento Helado']
}, index=['Pikachu', 'Pikachu', 'Charmander', 'Squirtle', 'Articuno'])
```

```python
df_estadisticas
```

```python
df_habilidades
```

## Inner join

Combina solo las filas que tienen índices coincidentes en ambos DataFrames o series, excluyendo aquellas filas donde no hay coincidencia en los índices.

```python
df_inner_join = df_estadisticas.join(df_habilidades, how='inner')
df_inner_join
```

## Right join

Combina las filas del DataFrame especificado (derecha) con las columnas del DataFrame llamante (izquierda) basándose en los índices.

```python
df_right_join = df_estadisticas.join(df_habilidades, how='right')
df_right_join
```

## Left join 

Combina las filas del DataFrame izquierda con las columnas del DataFrame o serie de la derecha basándose en los índices

```python
df_left_join = df_estadisticas.join(df_habilidades, how='left')
df_left_join
```

```python
df_unido = df_estadisticas.join(df_habilidades)
df_unido
```

## Outer join

Combina todas las filas de ambos DataFrames o series.

```python
df_outer_join = df_estadisticas.join(df_habilidades, how='outer')
df_outer_join
```
