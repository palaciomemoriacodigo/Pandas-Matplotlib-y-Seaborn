# Tipos de gráficos

## Gráficos univariados

### Histograma

Un histograma es un resumen de la distribución de un conjunto de datos numéricos, donde los datos se dividen en "bins", y la frecuencia de los datos se representa mediante barras.

$sns.histplot(data, x="variable")$

---

Parámetros:

**bins**: Controla el número de intervalos en los que se divide el rango de datos.

**binwidth**: Especifica el ancho de cada bin

**binrange**: Especifica el rango dentro del cual se calcularán los bins (por ejemplo, (inicio, fin)).

**color**: Define el color de las barras del histograma.

**fill**: Un booleano que indica si las barras deben estar rellenas o no.

**element**: Controla la forma de las barras ("bars", "step", "poly").

**stat**: Define la función de estadística a calcular. Por defecto es "count" que muestra la frecuencia. Otros valores pueden ser "frequency", "density", o "probability".
cumulative:



```python
sns.set_theme(style="darkgrid")
sns.histplot(data=df, x="age", kde=True, bins = 5, hue = 'sex', palette='bwr')
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.histplot(df['age'],kde=True)
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.histplot(df['age'],kde=True, color='c', fill=1, element='step', stat='probability')
plt.show()
```

### Displot

Permite dibujar distribuciones de datos, donde especificas el tipo de gráfico que deseas con el parámetro **kind**

---

**hue**: Variable en los datos que definirá los colores de las diferentes subgrupos.

**col y row**: Variables para hacer gráficos facetados (gráficos separados para cada categoría).

**kind**: El tipo de gráfico para dibujar ("hist", "kde", "ecdf").

**rug**: Si es True, añade "marcas de alfombra" en la parte inferior del gráfico para mostrar cada observación individual.

**kde**: Si es True y kind="hist", añade una línea KDE al histograma.

**bins**: Número de bins para el histograma, similar al parámetro en histplot.

**binwidth**: Ancho de cada bin para el histograma.

**stat**: La función estadística para calcular y mostrar en el histograma.

**element**: El elemento gráfico para usar en el histograma ("bars", "step", "poly").



Histograma `kde=True`

```python
sns.displot(df['age'], kind='hist', bins=20, kde=True)
plt.title('Histograma con KDE')
plt.show()
```

```python
sns.displot(df['age'], kind='kde')
plt.title('Estimación de Densidad Kernel')
plt.show()
```

### Diagrama de densidad

Es utilizado para visualizar la Estimación de Densidad de Kernel (KDE) de un conjunto de datos. 

$sns.kdeplot(data['variable']$

---


**cumulative**: Si es True, la KDE será acumulativa, lo que resulta en una función de distribución acumulativa.

**bw_adjust**: Factor que multiplica el estimador de ancho de banda utilizado en el cálculo de KDE. Un valor mayor produce una curva más suave, mientras que un valor menor produce una curva más rugosa.

**fill**: Si es True, el área bajo la curva KDE será rellenada.

**gridsize**: Número de puntos en la cuadrícula de evaluación. Un número mayor aumenta la resolución de la curva KDE.

**cut**: Número de desviaciones estándar pasadas las cuales se extiende la curva KDE; por defecto, es 3, lo que significa que la curva es visible tres desviaciones estándar más allá del soporte de los datos.

```python
plt.figure(figsize=(6,6))
sns.kdeplot(df['age'])
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.kdeplot(df['age'], fill=True, cut = 2, gridsize=10000)
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.kdeplot(df['age'], fill=True, cut = 2, gridsize=10000, cumulative=True)
plt.show()
```

### Diagrama de correlación

$sns.heatmap(data.corr())$

---

vmin, vmax: Valores mínimos y máximos del mapa de colores y la barra de color.

cmap: Mapa de colores para la visualización. Seaborn permite una variedad de mapas de colores de Matplotlib.

center: Un valor numérico que centra el mapa de colores. Útil cuando se desea enfatizar la divergencia alrededor de un punto central.

annot: Si es True, los valores serán escritos en las celdas del mapa de calor.

fmt: Formato de cadena para usar cuando annot es True.

linewidths: Ancho de las líneas que dividirán cada celda.

linecolor: Color de las líneas que dividen las celdas.

cbar: Si es True, muestra la barra de colores. A veces es útil ocultarlo para ahorrar espacio.

cbar_kws: Argumentos de palabras clave para ajustar la barra de colores.

square: Si es True, hace que cada celda del mapa de calor sea cuadrada.

```python
sns.heatmap(
    df.corr(numeric_only=True),
    linewidth=0.5,
    cmap='viridis',
    annot=True
).set_title('Heatmap')
plt.show()
```

```python
sns.heatmap(
    df.corr(numeric_only=True),
    linewidth=0.5,
    cmap='viridis',
    square=True,
    annot=False
).set_title('Heatmap')
plt.show()
```

```python
sns.heatmap(
    df.corr(numeric_only=True),
    linewidth=0.5,
    cmap='viridis',
    annot=True,
    square=True,
    fmt=".1f" 
).set_title('Heatmap')
plt.show()
```

### Facetgrid

Permite visualizar múltiples vistas de un conjunto de datos.

$sns.FacetGrid(data, col="variable")$

---

**hue**: Variable por la que se puede colorear. Esto añade una tercera dimensión de información sin tener que subdividir los gráficos.

```python
g = sns.FacetGrid(
    data=df, 
    col='class',        
    col_wrap=3,         
    height=4,          
    hue='class',        
    palette='viridis',  
    sharex=True,       
    sharey=True,      
    aspect=1,                 
)

g.map(sns.scatterplot, 'age', 'fare')
plt.subplots_adjust(top=0.9)
plt.show()
```

##  Gráficos categóricos

### Diagrama de barras

 Permtien una estimacion del valor central de una variable numérica con la altura de cada barra y proporciona una estimación de la incertidumbre usando barras de error.

$sns.barplot(data=data, x="category\_var", y="num\_var")$

---


**hue**: Variable categórica en data que se usa para producir barras de colores diferentes en el gráfico.

**order, hue_order**: Orden específico para plotear las categorías de x o hue.

**estimator**: Función estadística para estimar dentro de cada categoría categórica (por defecto es mean).

**errorbar**: Podemos añadir como parametro ci  (intervalo de confianza para las barras de error). El ci tiene como valor predeterminado es 95, 

(intervalo de confianza es del 95%).

**n_boot**: Número de iteraciones bootstrap para calcular el intervalo de confianza.

**units**: Identificador de muestreo para realizar un cálculo de error más detallado.

**orient**: Orientación de las barras ('v' para vertical y 'h' para horizontal). 

**color**: Color para todos los elementos, o una paleta si hue es usado.

**palette**: Colores a usar para los diferentes niveles de la variable hue.

**saturation**: Saturación del color de las barras.

**capsize**: Ancho de las "tapas" en las barras de error.

```python
sns.barplot(x='sex', y='survived', data=df, hue='sex')
plt.title('Gráfico de barras: Supervivencia por Género')
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.barplot(x='sex', y='survived', hue='class', data=df)
plt.title('Supervivencia por Género y Clase')
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.barplot(
    x='sex',
    y='survived',
    hue='class',
    data=df,
    estimator=np.mean,  
    errorbar=('ci',99.75),
    n_boot=1000, 
    units=None,  
    orient='v',  
    color=None,  
    palette='viridis', 
    saturation=0.75,  
    err_kws={'color': 'gray', 'linewidth': 2}, 
    capsize=0.05  
)

plt.title('Supervivencia por Género y Clase')  
plt.show() 
```

### Box plot

Permite visualizar distribuciones de datos a través de cuartiles. Indicando si una variable numérica difiere significativamente entre distintos grupos y permitiendo detectar valores atípicos.

**Mediana (segundo cuartil)**: La línea dentro de la caja que indica el valor medio de los datos.

**Cuartiles**: La caja en sí se extiende desde el primer cuartil (25%) hasta el tercer cuartil (75%) de los datos, proporcionando una vista del rango intercuartílico (IQR).

**Bigotes (whiskers)**: Las líneas que se extienden desde la caja indican la variabilidad fuera del cuartil superior e inferior. Se extienden hasta 1.5 * IQR desde los cuartiles.

**Valores atípicos**: Puntos individuales que ocurren fuera de los extremos de los bigotes se consideran valores atípicos y se trazan como puntos individuales.

$sns.boxplot(data=data, x="category\_var", y="num\_var")$

---

**hue**: Variable categórica en data que se usa para producir cajas de colores diferentes en el gráfico.

**order, hue_order**: Orden específico para plotear las categorías de x o hue.

**orient**: Orientación de las cajas ('v' para vertical y 'h' para horizontal).

**color**: Color para todos los elementos, o una paleta si hue es usado.

**palette**: Colores a usar para los diferentes niveles de la variable hue.

**width**: Ancho de las cajas, que es útil para comparar grupos que tienen diferentes tamaños de muestras.

**dodge**: Cuando hue está presente, dodge=True separa las cajas por hues; de lo contrario, las alinea.

**linewidth**: Ancho de las líneas que forman las cajas y los bigotes.

**whis**: Proporción del IQR más allá de la cual se consideran valores atípicos. Por defecto es 1.5.

```python
plt.figure(figsize=(6,6))
plt.style='whitegrid'
sns.boxplot(x="survived", y="age", hue = 'survived', data=df)
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.style='whitegrid'
sns.boxplot(x='day', y='tip', hue = 'day', data=tips_df)
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.boxplot(x='day', y='tip', hue='day', data=tips_df, palette='Set3', width=0.5, linewidth=2, dodge=False, whis=1.5)
plt.title('Distribución de propinas por día')
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.boxplot(x='day', y='tip', hue='day', data=tips_df, palette='Set3', width=0.5, linewidth=2, dodge=False, whis=2)
plt.title('Distribución de propinas por día')
plt.show()
```

### Pie chart

Se construye con Matplotlib, puedes ver como se distribuyen las categorías en un diagrama de tarta.

**labels**: Etiquetas para cada sector. Usualmente, se utilizan los nombres de las categorías.

**autopct**: Formato de cadena o función utilizada para etiquetar las cuñas con su valor numérico. El formato '%1.1f%%' muestra el porcentaje con un decimal.

**startangle**: Gira el inicio del gráfico de tarta en grados desde el eje x (90 grados coloca el primer sector en la parte superior).

**colors**: Lista de colores para las cuñas.

**shadow**: Si es True, dibuja una sombra debajo del gráfico.

**radius**: Radio del círcul

```python
class_counts = df['class'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(class_counts, labels=class_counts.index, autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff','#99ff99'])
plt.title('Distribución de pasajeros por clase en el Titanic')
plt.show()
```

### Violin plot

Muestra la densidad de los datos en diferentes valores, Combinando boxplots y gráficos de densidad de kernel. 

Una línea central muestra la distribución de los datos. La anchura más ancha indica una mayor densidad de puntos. Dentro podemos apreciar la mediana y los cuartiles

$sns.violinplot(data=data, x="category\_var", y="num\_var")$

---

**bw**: Ancho de banda para la estimación KDE, puede ser 'scott', 'silverman', un escalar o una función.

**cut**: Controla hasta dónde se extienden los extremos de los violines más allá de los valores extremos de los datos. El valor por defecto es 2.

**density_norm**: Método para escalar el ancho de los violines. Las opciones incluyen 'area', 'count', y 'width'; 'area' escala cada violín para tener el mismo área, 'count' los escala por el número de observaciones en esa categoría, 'width' usa el mismo ancho para todos.

**inner**: Representación de los datos cuantitativos dentro del violín, opciones incluyen 'box', 'quartile', 'point', 'stick', o None.

**split**: Si es True y hay dos niveles en hue, los violines se dibujarán como violines compartidos, lo que es útil para comparar directamente las distribuciones.

**palette**: Mapa de colores para los diferentes niveles de la variable hue.

**linewidth**: Ancho de las líneas del contorno del violín y del boxplot interno.

```python
sns.violinplot(x="survived", y="age", hue='sex', data=df)
plt.show()
```

```python
plt.figure(figsize=(10, 6)) 
sns.violinplot(
    x="survived",
    y="age",
    hue="sex",
    data=df,
    palette="Pastel1",  
    split=True,         
    inner="box",  
    density_norm="area",      
    linewidth=2        
)

plt.title('Distribución de Edad por Supervivencia y Género') 
plt.show()  
```

## Gráficos de regresión

### Regresión lineal

Permite trazar gráficos de regresión que permiten visualizar la relación lineal entre dos variables numéricas.

$sns.regplot(data=data, x="var1", y="var2")$

---

**color**: Color de la línea de regresión y de los puntos de dispersión.

**marker**: Formato de los marcadores de los puntos en el gráfico.

**scatter**: Booleano que indica si se deben dibujar los puntos de dispersión. Por defecto es True.

**fit_reg**: Booleano que indica si se debe calcular y dibujar la línea de regresión. Por defecto es True.

**ci**: Tamaño del intervalo de confianza para la estimación de la regresión, en porcentaje.

**n_boot**: Número de iteraciones bootstrap para calcular el intervalo de confianza.

**order**: Orden del polinomio para la regresión polinomial. Por defecto es 1, que es una regresión lineal.

**logx**: Si es True, el eje x se transformará en logarítmico.

**x_estimator**: Una función que colapsa x en una única estimación para cada valor único, útil para x categórico.

**x_bins**: Una forma de discretizar x y luego estimar y. Esto puede ser un número entero de bins, o un objeto array que define los bins.

```python
sns.regplot(x="age", y="fare", data=df, color='c')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.regplot(
    x="age",
    y="fare",
    data=df,
    color='c',            
    marker='o',          
    scatter_kws={"s": 50, "alpha": 0.5}, 
    line_kws={"lw": 2},      
)

plt.title('Relación entre Edad y Tarifa')  
plt.xlabel('Edad')  
plt.ylabel('Tarifa')  
plt.show() 
```

```python
slope, intercept = np.polyfit(df['age'].dropna(), df['fare'].dropna(), 1)
plt.figure(figsize=(10, 6))

ax = sns.regplot(
    x="age",
    y="fare",
    data=df,
    color='c',
    marker='o',
    scatter_kws={"s": 50, "alpha": 0.5},
    line_kws={"lw": 2}
)
slope, intercept = np.polyfit(df['age'].dropna(), df['fare'].dropna(), 1)
equation = f'y = {slope:.2f}x + {intercept:.2f}'

plt.text(0.05, 0.95, equation, transform=ax.transAxes, fontsize=12,
         verticalalignment='top', bbox=dict(boxstyle='round', facecolor='white', alpha=0.5))

plt.title('Relación entre Edad y Tarifa')
plt.xlabel('Edad')
plt.ylabel('Tarifa')
plt.show()
```

### Regresión con lmplot

Lmplot combina regplot y FacetGrid, lo que permite crear gráficos de regresión lineal con la facilidad de agregar múltiples gráficos basados en categorización

$sns.lmplot(data=data, x="var1", y="var2", hue="category")$

---

**hue**: Variable categórica en data que se usa para producir líneas de regresión de diferentes colores.

**col, row**: Variables para hacer gráficos facetados. Crea múltiples gráficos en una cuadrícula.

**palette**: Colores a usar para los diferentes niveles del hue.

**col_wrap**: Limita el número de gráficos por fila cuando se utiliza faceting por columnas.

**height, aspect**: Altura y relación de aspecto de cada gráfico, respectivamente.

**markers**: Estilos de los marcadores en el gráfico de dispersión.

**scatter_kws**, line_kws: Diccionarios adicionales para pasar parámetros a las funciones de gráficos de puntos y líneas, respectivamente.

**ci**: Tamaño del intervalo de confianza para la estimación de regresión.

**order**: Orden del polinomio en el ajuste de regresión polinomial.

```python
sns.lmplot(x="age", y="fare", hue="sex", data=df)
plt.show()
```

```python
sns.lmplot(
    x="age",
    y="fare",
    hue="sex",
    data=df,
    palette="Set1",  
    height=5,        
    aspect=1.5,     
    markers=['o', 'x'],  
    scatter_kws={"s": 50, "alpha": 0.7}, 
    line_kws={"lw": 2}   
)

plt.title('Relación entre Edad y Tarifa por Género')  
plt.show()
```

```python
g = sns.lmplot(
    x="age",
    y="fare",
    hue="sex",
    data=df,
    palette="Set1",
    height=5,
    aspect=1.5,
    markers=['o', 'x'],
    scatter_kws={"s": 50, "alpha": 0.7},
    line_kws={"lw": 2}
)

for ax, sex in zip(g.axes.flat, df['sex'].unique()):
    subset = df[df['sex'] == sex]
    slope, intercept = np.polyfit(subset['age'], subset['fare'], 1)
    equation = f'y = {slope:.2f}x + {intercept:.2f}'
    ax.text(0.05, 0.95, equation, transform=ax.transAxes, fontsize=12,
            verticalalignment='top', bbox=dict(boxstyle='round', facecolor='white', alpha=0.5))

plt.title('Relación entre Edad y Tarifa por Género')
plt.show()
```

## Gráficos de estructura compleja

### Cluster map

Permite visualizar una matriz de datos como un mapa de calor junto con un dendrograma, que agrupa las filas y columnas basadas en la similitud de sus datos.

$sns.clustermap(data.corr())$

---

**pivot_kws**: Diccionario de argumentos para pivot cuando data necesita ser reformada antes de visualización.

**method**: Método de agrupación jerárquica, por ejemplo, 'single', 'complete', 'average', 'ward', etc.

**metric**: Métrica de distancia, por ejemplo, 'euclidean', 'minkowski', 'cityblock', etc.

**z_score**: Eje para el cual calcular la puntuación Z (resta la media y divide por la desviación estándar).

**standard_scale**: Eje para el cual normalizar los datos restando el valor mínimo y dividiendo por el rango máximo.

**figsize**: Tamaño de la figura.

**cmap**: Mapa de colores para visualizar los elementos del mapa de calor.

**linewidths**: Ancho de las líneas que separan las celdas del mapa de calor.

**linecolor**: Color de las líneas que separan las celdas.

**cbar_kws**: Diccionario de argumentos para configurar la barra de color.

**row_cluster, col_cluster**: Booleano para controlar si las filas o columnas deben ser agrupadas.

**annot**: Booleano o rectangular dataset. Si es True, escribe el dato en cada celda.

**annot_kws**: Diccionario de argumentos para las anotaciones de texto.

```python
dummies = pd.get_dummies(tips_df)
corr_encoded = dummies.corr()
sns.clustermap(corr_encoded, annot=True, cmap='coolwarm', linewidths=.5)
plt.show()
```

```python
tips_df = sns.load_dataset('tips')
tips_df.drop(['day', 'sex'], axis=1, inplace=True)
dummies = pd.get_dummies(tips_df)
corr_encoded = dummies.corr()

sns.clustermap(
    corr_encoded,
    annot=True,            
    cmap='coolwarm',       
    linewidths=.5,         
    figsize=(10, 10),      
    row_cluster=True,      
    col_cluster=True,      
    cbar_kws={'shrink': .5} 
)

plt.show()
```

## Gráficos especiales

### Strip plot

Permite representar una distribución de datos en categorías con una disposición de puntos. Es similar a un "scatter plot" pero para una variable categórica. 

$sns.stripplot(data=data, x="category", y="num\_var")$

---

**hue**: Variable en data que determina el color de los puntos de datos, proporcionando una capa adicional de agrupación.

**order, hue_order**: Especifica un orden para las categorías y los colores.

**jitter**: Si es True o un float, añade ruido a las coordenadas de categoría para evitar la superposición de puntos, mejorando la legibilidad.

**dodge**: Si es True, separa los puntos por hue para evitar la superposición cuando se usa hue.

**orient**: Orientación de los puntos ('v' para vertical y 'h' para horizontal).

**color**: Color para todos los elementos, o una paleta si hue es usado.

**palette**: Mapa de colores para los diferentes niveles de la variable hue.

**size, linewidth**: Propiedades para ajustar el tamaño y el ancho de línea de los puntos.

**marker**: Estilo del marcador utilizado para cada punto.

```python
plt.figure(figsize=(8,8))
sns.stripplot(x="sex", y="age",  hue='sex', data=df)
plt.show()
```

```python
plt.figure(figsize=(8, 8))  

sns.stripplot(
    x="sex", 
    y="age", 
    hue="sex", 
    data=df,
    jitter=True, 
    dodge=True,   
    palette="Set2",  
    marker="D",   
    linewidth=1  
)

plt.title('Distribución de Edad por Género') 
plt.show()  
```

### Swarm plot

 Permite visualizar la distribución de los datos de manera que cada punto de datos individual sea visible.  Es similar al stripplot, pero los puntos se ajustan de modo que no se superponen.

$sns.swarmplot(data=data, x="category", y="num\_var")$

---

**hue**: Variable categórica que determina el color de los puntos de datos.

**data**: DataFrame de pandas que contiene las variables.

**order, hue_order**: Orden específico para las categorías de x o hue.

**dodge**: Si es True, separa los puntos por hue para evitar la superposición cuando se utiliza hue.

**orient**: Orientación de los puntos ('v' para vertical y 'h' para horizontal).

**color, palette**: Controla el color de los puntos.

**size**: Tamaño de los puntos en el gráfico.

**edgecolor**: Color del borde de cada punto.

**linewidth**: Ancho de las líneas del borde de cada punt

```python
plt.figure(figsize=(10, 6))
sns.swarmplot(x='class', y='age', hue = 'class', data=df, size=3)
plt.title('Swarm Plot')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.swarmplot(
    x='class', 
    y='age', 
    data=df, 
    size=4,  
    color='blue',  
    edgecolor='black', 
    linewidth=1 
)
plt.title('Distribución de Edad por Clase') 
plt.show() 
```
